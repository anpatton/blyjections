```python
import polars as pl
from pybaseball import chadwick_register
```

```python

# Option 1: Get all player IDs from Chadwick Register
def get_all_player_ids(min_year: int = 2022):
    """Get all player IDs for players active after a given year."""
    
    # Get complete player register
    players = chadwick_register()
    
    # Filter for players active after min_year
    recent_players = players[players['mlb_played_last'] >= min_year]
    recent_players = pl.from_pandas(recent_players)
    
    return recent_players

# Usage
players_df = get_all_player_ids()
```
